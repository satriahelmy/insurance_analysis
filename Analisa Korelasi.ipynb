# Analisa Korelasi

Pada tahapan ini, kita akan mengecek hubungan antara 2 variable dengan menggunakan korelasi.

## Import Data

```python
import pandas as pd
import numpy as np

insurance = pd.read_csv("data/insurance.csv")
insurance
```

## Hubungan antara tagihan dengan BMI

```python
corr_tagihan_bmi = insurance[['bmi','charges']].corr()
corr_tagihan_bmi
```

## Hubungan antara tagihan dengan umur

```python
corr_tagihan_umur = insurance[['age','charges']].corr()
corr_tagihan_umur
```

## Hubungan antara tagihan dengan jumlah anak

```python
corr_tagihan_children = insurance[['children','charges']].corr()
corr_tagihan_children
```

## Hubungan antara tagihan dengan perokok

```python
def check_is_smoker(row):
    if row['smoker'] == 'yes':
        return 1
    else:
        return 0

insurance['is_smoker'] = insurance.apply(lambda row: check_is_smoker(row), axis=1)
insurance
```

```python
corr_tagihan_smoker = insurance[['is_smoker','charges']].corr()
corr_tagihan_smoker
```

## Hubungan antara tagihan dengan jenis kelamin

```python
def check_jenis_kelamin(row):
    if row['sex'] == 'male':
        return 1
    else:
        return 0

insurance['jenis_kelamin'] = insurance.apply(lambda row: check_jenis_kelamin(row), axis=1)
insurance
```

```python
corr_tagihan_jk = insurance[['jenis_kelamin','charges']].corr()
corr_tagihan_jk
```
